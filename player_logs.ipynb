#Random Stats

```python
ppg = df.groupby('PLAYER_NAME')['PTS'].mean().sort_values(ascending=False).head(10)

#Top 10 Players by Points visual

ppg.plot(kind='barh', color='skyblue')
plt.title('Top 10 Players by Points Per Game')
plt.xlabel('Points Per Game')
plt.ylabel('Player')
plt.show()
```

```python
#top ten rebounds per game
rpg = df.groupby('PLAYER_NAME')['REB'].mean().sort_values(ascending=False).head(10)

#Top 10 Players by Rebounds Per Game visual

rpg.plot(kind='barh', color='skyblue')
plt.title('Top 10 Players by Rebounds Per Game')
plt.xlabel('Rebounds Per Game')
plt.ylabel('Player')
plt.show()
```

```python
#Top 10 3pt makes and percentage

three_pt_makes = df.groupby('PLAYER_NAME')['FG3M'].sum().sort_values(ascending=False).head(10)

three_pt_makes.plot(kind='barh', color='skyblue')
plt.title('Top 10 Players by 3pt Makes')
plt.xlabel('3pt Makes')
plt.ylabel('Player')
plt.show()

#Top 10 3pt percentage
three_pt_percentage = df.groupby('PLAYER_NAME')['FG3_PCT'].mean().sort_values(ascending=False).head(10)

three_pt_percentage.plot(kind='barh', color='skyblue')
plt.title('Top 10 Players by 3pt Percentage')
plt.xlabel('3pt Percentage')
plt.ylabel('Player')
plt.show()

#Top 10 3pt percentage
```

Do Three point make avgs correlate with percentage negative or positively

```python
#correlation between 3pt makes per game and 3pt percentage 

corr = df[['FG3M', 'FG3_PCT']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation between 3pt makes per game and 3pt percentage')
plt.show()
```
